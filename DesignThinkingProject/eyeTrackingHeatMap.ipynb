```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

points = [
    (300, 400), (305, 405), (310, 410), (800, 200),
    (850, 210), (900, 215), (1200, 700), (1250, 710),
    (1300, 720), (1000, 540), (1020, 550), (1040, 530)
]

df = pd.DataFrame(points, columns=["x", "y"])
heatmap = np.zeros((1080, 1920))
for x, y in points:
    try:
        heatmap[int(y), int(x)] += 1  
    except IndexError:
        pass  

blurred_map = cv2.GaussianBlur(heatmap, (99, 99), 0)
plt.figure(figsize=(10, 5))
plt.imshow(blurred_map, cmap='Accent', interpolation='mitchell')
plt.title('Eye Tracking Heatmap Simulated Data')
plt.axis('off')
plt.colorbar(label='Intensity')
plt.tight_layout()
plt.show()
```
