```python
html_code = '''<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>Mental Health Chatbot</title>
  <style>
    body {
      font-family: Arial, sans-serif;
      background-color: #f4f6f8;
      padding: 20px;
      display: flex;
      justify-content: center;
    }
    .chat-container {
      width: 100%;
      max-width: 600px;
      background-color: #ffffff;
      border-radius: 10px;
      box-shadow: 0 4px 10px rgba(0, 0, 0, 0.1);
      padding: 20px;
    }
    .chat-box {
      height: 400px;
      overflow-y: auto;
      display: flex;
      flex-direction: column;
      margin-bottom: 15px;
      border: 1px solid #ccc;
      border-radius: 8px;
      padding: 10px;
    }
    .message {
      margin: 8px 0;
      padding: 10px 15px;
      border-radius: 15px;
      max-width: 75%;
      font-style: italic;
    }
    .user {
      background-color: #d3f9d8;
      align-self: flex-end;
    }
    .bot {
      background-color: #d0ebff;
      align-self: flex-start;
    }
    .input-area {
      display: flex;
    }
    input {
      flex: 1;
      padding: 10px;
      font-size: 16px;
      border-radius: 8px;
      border: 1px solid #ccc;
    }
    button {
      margin-left: 10px;
      padding: 10px 20px;
      font-size: 16px;
      border: none;
      border-radius: 8px;
      background-color: #3498db;
      color: white;
      cursor: pointer;
    }
    select {
      margin-bottom: 15px;
      padding: 8px;
      font-size: 16px;
      border-radius: 6px;
    }
  </style>
</head>
<body>
  <div class="chat-container">
    <h2>Mental Health Chatbot</h2>
    <select id="modeSelect" onchange="switchMode()">
      <option value="positive">😊 Positive Conversation</option>
      <option value="supportive">😔 Supportive Conversation</option>
    </select>
    <div id="chatBox" class="chat-box"></div>
    <div class="input-area">
      <input type="text" id="userInput" placeholder="Type your message...">
      <button onclick="sendMessage()">Send</button>
    </div>
  </div>

  <script>
    const positiveReplies = [
      "Hi, how are you feeling today?",
      "That's wonderful to hear! What's been the highlight of your day?",
      "That sounds amazing. It's great to enjoy the little things.",
      "Keep embracing those happy moments! They're important.",
      "Thanks for sharing such good vibes. It was nice talking to you. 😊"
    ];

    const supportiveReplies = [
      "Hi, how are you feeling today?",
      "I'm really sorry you're going through that. Would you like to talk more about it?",
      "It's completely okay to feel this way. You're not alone.",
      "You're doing your best, and that's enough. I'm here with you.",
      "Thank you for trusting me with how you feel. 💛"
    ];

    let positiveIndex = 0;
    let supportiveIndex = 0;
    let currentMode = "positive";

    function switchMode() {
      currentMode = document.getElementById("modeSelect").value;
      document.getElementById("chatBox").innerHTML = "";
      document.getElementById("userInput").value = "";
      positiveIndex = 0;
      supportiveIndex = 0;
    }

    function sendMessage() {
      const input = document.getElementById("userInput");
      const message = input.value.trim();
      if (message === "") return;

      addMessage(message, "user");
      input.value = "";

      setTimeout(() => {
        let reply = "";
        if (currentMode === "positive" && positiveIndex < positiveReplies.length) {
          reply = positiveReplies[positiveIndex++];
        } else if (currentMode === "supportive" && supportiveIndex < supportiveReplies.length) {
          reply = supportiveReplies[supportiveIndex++];
        } else {
          reply = "I'm here whenever you're ready to continue. 💬";
        }
        addMessage(reply, "bot");
      }, 600);
    }

    function addMessage(text, sender) {
      const chatBox = document.getElementById("chatBox");
      const messageDiv = document.createElement("div");
      messageDiv.className = "message " + sender;
      messageDiv.textContent = text;
      chatBox.appendChild(messageDiv);
      chatBox.scrollTop = chatBox.scrollHeight;
    }

    document.getElementById("userInput").addEventListener("keypress", function(e) {
      if (e.key === "Enter") {
        sendMessage();
      }
    });
  </script>
</body>
</html>'''

with open("mental_health_chatbot_final.html", "w", encoding="utf-8") as f:
    f.write(html_code)

import webbrowser
webbrowser.open("mental_health_chatbot_final.html")
```
